# Matplotlib Notes
- Matplotlib is the most popular library fordata visualization in python
- works well with Numpy and Pandas

```python
#import library
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Basic plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x,y)
plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Exercise
# Plot y = x for x = 0 to 10.
x = [i for i in range(11)]
y = [i**2 for i in x]

plt.plot(x, y)
plt.title('y = x ^ 2')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Styling the plot
plt.plot(x, y, color='red', linestyle="--", marker="o")
plt.show()
```

```python
#Exercise 
#Plot sin(x) and cos(x) with different colors and markers.
x = np.linspace(0,2*np.pi,100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, color='blue' , marker='o' , label='sinx')
plt.plot(x, y_cos, color='red', marker='s', label='cosx')

plt.title("Sine and Cosine Functions")
plt.xlabel('x values')
plt.ylabel('Function values')
plt.legend()
```

```python
# Bar Chart

categories = ["A", "B", "C"]
values = [10, 20, 15]

plt.bar(categories, values)
plt.title('Bar Chart')
plt.show()
```

```python
# Exercise
# Create a bar chart showing sales for 4 products.
product = ['shoes','skirt','shirt','heels']
sales = [100,200,300,50]

plt.bar(product, sales)
plt.title('Sales for products')
plt.show()
```

```python
# Scatter plot
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.show()
```

```python
# Histogram
data = np.random.randn(1000)

plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title('Histogram')
plt.show()
```

```python
# Exercise
# Plot a histogram of 500 random numbers from a normal distribution.
data = np.random.randn(500)

plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram of 500 random numbers from normal distribution.')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Multiple plots(Subplots)
x= np.linspace(0,10,100)
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
plt.title('Cosine')

plt.tight_layout()
plt.show()
```
