# Seaborn Note
- Built on top of Matplotlib.

- Makes statistical plots easier and prettier (colors, themes, automatic grouping).

- Integrates directly with Pandas DataFrames.

## Why Seaborn?
- Fewer lines of code for advanced plots.

- Great for exploratory data analysis (EDA).

- Common in finance, health, and research because it can:

    - Visualize distributions (hist, KDE, boxplot, violinplot).

    - Compare categories (bar, count, swarm).

    - Show correlations (heatmap, pairplot).

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load example dataset
tips = sns.load_dataset("tips")
tips.head()
```

# Basics Plots

```python
# ScatterPlot

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
#Line Plot
sns.lineplot(x='size', y='total_bill', data=tips)
plt.show()
```

# Distribution Plots

```python
# Histogram
sns.histplot(tips['total_bill'], kde=True)
plt.show()
```

```python
# BoxPlot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
# Violin Plot
sns.violinplot(x='day', y='tip', data=tips)
plt.show()
```

# Categorical Plots

```python
sns.barplot(x='day', y='total_bill', data=tips)
plt.show()
```

# Correlation & Heatmaps

```python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

# Pairplot (EDA shortcut)



```python
sns.pairplot(tips, hue='sex')
plt.show()
```
